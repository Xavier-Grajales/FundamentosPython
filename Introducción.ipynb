# Operadores lógicos.
Algunos de los operadores relacionales comunes son:


image.png

Los operadores relacionales dan como resultado TRUE o False, es probable que sea necesario recordar lo siguiente:
image.png


```python
#Definimos algunos datos para poder ejemplficar el uso de and
edad=18
estatura=1.80
'''
En este caso definiremos utilizaremos operadores relaciones y operadores lógicos 
preguntaremos si la edad es igual a 18 y si la altura debe es mayor a 1.70
como esto se cumplirá debemos obtener un resultado TRUE si alguna 
de las condiciones no se cumplierá tendriamos un false

'''
test= (edad==18) and (estatura>1.70)
print(test)

'''
Ahora utilizaremos un or y nos damos cuenta que si al menos una de las condiciones
se cumple tenemos TRUE 
'''
test= (edad==16) or (estatura>1.70)
print(test)
```
